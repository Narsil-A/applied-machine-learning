Question 6

Using your knn classifier, predict the class label using the mean value for each feature.

```python
def answer_six():
    # YOUR CODE HERE

    cancer_df = answer_one()

    # Calculate the mean of each feature, excluding the target
    feature_means = cancer_df.drop('target', axis=1).mean().values.reshape(1, -1)

    # Retrieve the trained KNN classifier
    knn = answer_five()

    means_df = pd.DataFrame(feature_means, columns=cancer_df.columns[:-1])

    # Use the classifier to predict the class label for the mean feature values
    prediction = knn.predict(means_df)

    return prediction

    raise NotImplementedError()


print(answer_six())
```

Try using the plotting function below to visualize the different predicition scores between train and test sets, as well as malignant and benign cells.
